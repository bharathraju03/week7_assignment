```python
import pandas as pd

df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")

df.head()
```

```python
print("First 5 rows:")
display(df.head())

print("Last 5 rows:")
display(df.tail())

print("Shape:")
print(df.shape)

print("Columns:")
print(df.columns)

print("Data Types:")
print(df.dtypes)
```

```python
print("Missing Values:")
print(df.isnull().sum())

df = df.fillna(0)

print("\nAfter Filling:")
print(df.isnull().sum())
```

```python
df[df["Sales"] > 1000]
```

```python
df[["Product Name", "Sales"]]
```

```python
print("Shape before removing duplicates:")
print(df.shape)

df = df.drop_duplicates()

print("Shape after removing duplicates:")
print(df.shape)
```

```python
df["total_amount"] = df["Sales"] * df["Quantity"]

df[["Sales", "Quantity", "total_amount"]].head()
```

```python
df.to_csv("cleaned_superstore.csv", index=False)

print("Cleaned dataset saved successfully!")
```
